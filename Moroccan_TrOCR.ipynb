### load the TRocr model From hugging Face

````python
model_name = "microsoft/trocr-base-printed"
processor = TrOCRProcessor.from_pretrained(model_name)
processor.feature_extractor.do_rescale = False  # not rescale pic since they are allready rescaled
model = VisionEncoderDecoderModel.from_pretrained(model_name)

#demander par Trocr pour le finetunning
# Set the decoder_start_token_id
model.config.decoder_start_token_id = processor.tokenizer.cls_token_id
model.gradient_checkpointing_enable()
 #Set the pad_token_id to the EOS token ID if it's not already set
# This is commonly used for text generation tasks
if model.config.pad_token_id is None:
    model.config.pad_token_id = processor.tokenizer.eos_token_id

# to fix the out of memory problem
# torch.cuda.empty_cache()
````

````python
transform = transforms.Compose([
    transforms.Resize((384, 384)),
    transforms.ToTensor(),
    # transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5]), #the pictures are already scaled
    # transforms.Lambda(lambda image: torch.clamp(image, 0, 1))
])

#  Dataset personnalisé
class OCRDataset(Dataset):
    def __init__(self, dataframe, processor, transform=None):
        self.dataframe = dataframe
        self.processor = processor
        self.transform = transform

    def __len__(self):
        return len(self.dataframe)

    def __getitem__(self, idx):
        row = self.dataframe.iloc[idx]
        image_path = row["Image_Path"]
        annotation = row["Annotation "]

        # Charger et transformer l'image
        image = Image.open(image_path).convert("RGB")
        # Get maximum pixel value for each channel
        max_pixel_values = image.getextrema()[1]

         # Check if max pixel value is greater than 1
        if isinstance(max_pixel_values, tuple) and any(val > 1 for val in max_pixel_values):
            image = self.transform(image)  # Rescale only if needed
        else:
            image = transforms.ToTensor()(image)

        # Prétraiter les données pour le modèle
        encoding = self.processor(
            images=image,
            text=annotation,
            return_tensors="pt",
            padding="max_length",
            truncation=True,
            max_length=MAX_LENGTH
        )
        encoding["labels"] = encoding["labels"].squeeze()
        return encoding
````

````python
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
BATCH_SIZE = 30
EPOCHS = 20
LEARNING_RATE = 5e-5
````

````python
# Charger les fichier excel pour l'entrainement et l'evaluation
train_df = pd.read_excel("/content/drive/MyDrive/MRocr/Train_MRocr.xlsx")
valid_df = pd.read_excel("/content/drive/MyDrive/MRocr/Valid_MRocr.xlsx")
test_df = pd.read_excel("/content/drive/MyDrive/MRocr/Test_MRocr.xlsx")

# preparer les datasets
train_dataset = OCRDataset(train_df, processor, transform)
valid_dataset = OCRDataset(valid_df, processor, transform)
test_dataset = OCRDataset(test_df, processor, transform)

#DataLoaders
train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
valid_loader = DataLoader(valid_dataset, batch_size=BATCH_SIZE, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False)
````

````python
# 5. Configurer l'optimiseur
optimizer = AdamW(model.parameters(), lr = LEARNING_RATE)
model.to(DEVICE)
# 6. Fonction d'entraînement
def train_model(model, train_loader, valid_loader, optimizer, epochs):
    for epoch in range(epochs):
        model.train()
        total_loss = 0
        print(f"Epoch {epoch + 1}/{epochs}")

        for batch in train_loader:
            torch.cuda.empty_cache()
            pixel_values = batch["pixel_values"].squeeze().to(DEVICE)
            labels = batch["labels"].to(DEVICE)

            # Passer en avant et calculer la perte
            outputs = model(pixel_values=pixel_values, labels=labels)
            loss = outputs.loss
            total_loss += loss.item()

            # Backpropagation
            loss.backward()
            optimizer.step()
            optimizer.zero_grad()

        avg_train_loss = total_loss / len(train_loader)
        print(f"Training Loss: {avg_train_loss:.4f}")

        # Évaluer sur les données de validation
        evaluate_model(model, valid_loader)

        # Sauvegarder le modèle
        model.save_pretrained(f"trocr_finetuned_epoch_{epoch + 1}")
        processor.save_pretrained(f"trocr_finetuned_epoch_{epoch + 1}")

# 7. Fonction d'évaluation avec CER
def evaluate_model(model, data_loader):
    model.eval()
    total_loss = 0
    cer_values = []

    with torch.no_grad():
        for batch in data_loader:
            pixel_values = batch["pixel_values"].squeeze().to(DEVICE)
            labels = batch["labels"].to(DEVICE)

            # Générer les prédictions
            outputs = model.generate(pixel_values)
            predictions = processor.batch_decode(outputs, skip_special_tokens=True)
            references = processor.batch_decode(labels, skip_special_tokens=True)

            # Calculer la perte
            outputs = model(pixel_values=pixel_values, labels=labels)
            loss = outputs.loss
            total_loss += loss.item()

            # Calculer le CER pour chaque paire (prédiction, référence)
            for ref, pred in zip(references, predictions):
                cer_values.append(cer(ref, pred))

    avg_loss = total_loss / len(data_loader)
    avg_cer = sum(cer_values) / len(cer_values)
    print(f"Validation Loss: {avg_loss:.4f}, CER: {avg_cer * 100:.2f}%")
````

````python
import evaluate

cer_metric = evaluate.load("cer")
wer_metric=evaluate.load("wer")
bleu_metric = evaluate.load("bleu")


test_df = pd.read_excel("/content/drive/MyDrive/MRocr/Test_MRocr.xlsx")
predicted_texts = []
true_texts = test_df["Annotation "].tolist()  # Remplace "true_text" par le nom correct de la colonne des labels


for img_path in test_df["Image_Path"]:  # Remplace "image_path" par le bon nom de colonne
    try:
        image = Image.open(img_path).convert("RGB")
        pixel_values = processor(images=image, return_tensors="pt").pixel_values

        with torch.no_grad():
            generated_ids = model.generate(pixel_values)

        predicted_text = processor.batch_decode(generated_ids, skip_special_tokens=True)[0]
        predicted_texts.append(predicted_text)

    except Exception as e:
        print(f"Erreur lors du traitement de {img_path}: {e}")
        predicted_texts.append("")  # Ajouter un texte vide en cas d'erreur

# Calcul dles metriques d'evaluations
## cer
cer_score = cer_metric.compute(predictions=predicted_texts, references=true_texts)
## WER
wer_score = wer_metric.compute(predictions=predicted_texts, references=true_texts)
## Bleu
bleu_score = bleu_metric.compute(predictions=predicted_texts, references=true_texts)

````

````python
#les resultats d'evaluation
print("Character Error Rate (CER):", cer_score)
print("Word Error Rate (WER):", wer_score)
print("BLEU Score:", bleu_score)

````
